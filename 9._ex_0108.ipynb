### 1. MultiIndex:

```python
import pandas as pd
import numpy as np
import os
```

#### 1.1. Creating a DataFrame with MultiIndex:

```python
my_header = ['A','B','C']
```

```python
my_index_out = ['G1']*2 + ['G2']*2 + ['G3']*2
my_index_in = ['a', 'b']*3
```

```python
my_index_zipped = list(zip(my_index_out, my_index_in))
my_index_zipped
```

```python
my_multi_index = pd.MultiIndex.from_tuples(my_index_zipped)
df = pd.DataFrame(data=np.random.randn(6,3),index=my_multi_index,columns=my_header)
```

```python
df
```

```python
# Indexing followed by another indexing.
df.loc['G1'].loc['a']
```

```python
# Indexing followed by another indexing.
df.loc['G1'].loc['a','B']
```

#### 2.3. Pivoting:

```python
df = pd.read_csv('./data/weather.csv')
df
```

```python
df1=df.pivot(index='date',columns='city')
df1
# we try to make our analyst axes columns (by date, by city...)

```

```python
df1.columns
```

```python
df2=df.pivot(index='date',columns='city', values=['temperature'])
df2
```

```python
df3=df.pivot(index='date',columns='city', values=['temperature','humidity'])
df3
```

**Pivot table** is used to summarize and aggregate data inside dataframe.

```python
df.pivot_table(index='city',columns='date',aggfunc='sum')
```

```python
df.pivot_table(index='city',columns='date')
```

```python
df.pivot_table(index='city',columns='date',aggfunc=['sum','mean'])
```

```python
df.pivot_table(index='city',columns='date',aggfunc=['sum','mean'],margins=True)
```
