```python
import pandas as pd
```

# checking for missing values

```python
# read titanic dataset
df=pd.read_csv("./data/titanicPassengers.csv", sep=";")
df
```

```python
# Calculate the frequency of each category in column 'deck', including nan  (check value_counts)
df['Cabin'].value_counts()
```

```python
# find the missing values using is_null() method
df.isnull()
```

```python
# find the missing values using notnull() method
df.notnull()
```

```python
# calculate the number of the missing data using is_null() method
df.isnull().sum()
```

```python
# delete all columns with more than 500 NaN values 
df2=df.copy()
df2=df2.dropna(axis=1, thresh=500)
```

```python
# limit the subset of the column 'age'
df2.dropna(subset=['Age'])
```

```python
df2.isnull().sum()
```

```python
# replace the missing values with mean
df.fillna(df['Age'].mean(), inplace=True)
df.isnull().sum()
```

```python
# replace the missing values with mode
df.fillna(df['Age'].mode()[0], inplace=True)
df
```

# Checking and Processing Duplicate Dat

```python
import pandas as pd
df= pd.DataFrame({
    'c1':['a','a','b','a','b'],
    'c2':[1,1,1,2,2],
    'c3':[1,1,2,2,2]
})
df
```

```python
# check for duplicate data
df.duplicated()
```

```python
#find duplicate data in the column 'c2'
df['c2'].duplicated()
```

```python
# Remove duplicate rows based on column 'c2' and 'c3'
df.drop_duplicates(subset=['c2', 'c3'])
```
