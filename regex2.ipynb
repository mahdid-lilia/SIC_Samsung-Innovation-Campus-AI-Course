```python
# Read the book.
f = open("/content/drive/MyDrive/data_set/book_crime_and_punishment.txt",'r',encoding='UTF8')
book = f.readlines()
f.close()
```

```python
#  A function that returns the strings where the pattern is matched.
# The pattern is input as a regular expression.
def grep(reg_exp, my_text):
    result = []
    reg_exp_compiled = re.compile(reg_exp)
    for i in range(len(my_text)):
        if reg_exp_compiled.search(my_text[i]):
            result.append(my_text[i])
    return result
```

#### 1.2. Apply the grep() function:

Some tests.

Search the book "Crime and Punishment".

```python
# Sentences that begin with "This"
grep('^This', book)
```

```python
   # Sentences that end with "end."
grep("end.$",book)
```

```python
    # Sentences that begin with 'I' or 'i'
grep("^I | ^i", book)
```

```python
# Sentences that begin with a number.
grep("^\d", book)
```

```python
# Sentences with the pattern 'b.d'
grep("b.d", book)
```

```python
grep("ed | red", book)
```

```python
# Sentences with "love" or "hate".
grep("love|hate", book)
```

```python
# Percentage of the sentences with "crime" or "punishment".
100 * len(grep("crime|punishment",book))/len(book)
```
