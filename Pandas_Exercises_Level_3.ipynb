```python
import pandas as pd
```

```python
df_a = pd.DataFrame({'Name':['Harry Potter','David Baker', 'John Smith', 'Juan Martinez','Jane Connor'],
                     'Gender':['Male','Male','Male','Male','Female'],
                     'Age':[23,31,22,36,30]})
df_a
```

```python
df_b = pd.DataFrame({'Name':['Harry Potter','Alex Du Bois', 'Joanne Rowling','Jane Connor'],
                     'Position':['Intern','Team Lead','Manager','Manager'],
                     'Wage':[25000,75000,90000,70000]})
df_b
```

```python
# Merging DataFrames : A.name B.name used as key
# inner join 
# Inner join is the most common type of join youâ€™ll be working with. It returns a dataframe with only those rows that have common characteristics.
df_a.merge(df_b, left_on='Name', right_on='Name')
```

```python
df_a.merge(df_b, how='inner')
```

```python
#left join
df_a.merge(df_b, how='left')
```

```python
#right join
df_a.merge(df_b, how='right')
```

```python
# Full outer join
#Full Join, also known as Full Outer Join, returns all those records which either have a match in the left or right dataframe.
df_a.merge(df_b, how='outer')
```

```python
# concat df_A and df_B VERTICALLY
pd.concat([df_a,df_b])
```

```python
# concat df_A and df_B HORIZONTALLY
pd.concat([df_a,df_b], axis=1)
```

```python
df1 = pd.DataFrame({'a':['a0','a1','a2','a3'],
                     'b':['b0','b1','b2','b3'],
                     'c':['c0','c1','c2','c3']},
                      index=[0,1,2,3])
df1
```

```python
# concat df1, df2

pd.concat([df_a,df_b])
```

```python
# concat df1, df2, use the parameter 'ignore_index'
pd.concat([df_a,df_b],  ignore_index=True)
```

```python
# concat df1, df2, use the parameter 'axis=1'
pd.concat([df_a,df_b], axis=1)
```

## DataFrame Sorting and Multi-Index
