```python
data = pd.read_csv(
  "data/multivariate.csv", 
    sep=';'
)
```

```python
data
```

```python
# plot the distribution of Birth rate and Mortality rate in the same plot 
import seaborn as sns
sns.set_color_codes() 
sns.distplot(data['Birth rate'])
sns.distplot(data['Mortality rate'])
```

```python
sns.distplot(data['Mortality rate'],kde=True, hist=False)
```

```python
# print the correlation matrix 
mat=data.corr()
mat
```

```python
# plot the heatmap using the correlationn matrix
sns.set(rc = {'figure.figsize':(15,8)})
sns.heatmap(data.corr(),annot=True)
```

```python
# using pivot_table , average,min,max( morality_rate, birth_rate ) for each continent  , comments 
import numpy as np
table = pd.pivot_table(data, values=['Mortality rate', 'Birth rate'], index=['Continent'],
                      aggfunc=[np.mean,np.min, np.max])
table
```

```python
# Average pop and growth_rate  for each group in one code sentence. Use the groupby() method. 
data.groupby('Group')['Pop','Growth rate'].mean()
```

```python
# Sort the DataFrame in descending order of the growth_rate
data.sort_values(by=['Growth rate'],ascending=False)
```

```python
# get the countries where less than 5
data[data['Growth rate']<5]
```
